На всякий случай, сравним объем господдержки и бюджет фильма. Исходя из здравого смысла, бюджет не должен быть меньше суммы возвратной и невозвратной господдержки. 

```python
df['total_support'] = df['refundable_support'] + df['nonrefundable_support'] # посчитаем общую сумму господдержки
```

```python
df['budget_support_compare'] = df['total_support'] > df['budget'] # посмотрим, есть ли строки, где бюджет меньше, чем господдержка
```

```python
df['budget_support_compare'].unique() # посмотрим, сколько получилось таких значений
```

```python
df.loc[df['budget_support_compare'] == True] # посмотрим, в каких строках господдержка превышает бюджет. 
```

Мы видим, что часть значений в столбце budget равна нулю, хотя есть данные о том, что фильму была выделена господдержка. Заменим значения в столбце budget на значения из total_support

```python
df.loc[df['budget'] == 0] = df.loc[df['budget'] == 0].replace(df['total_support']) # заменим нулевые значения в budget на total_support
```

```python
df['budget'].value_counts() # проверим результат
```
