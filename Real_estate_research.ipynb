```python
df['ceiling_height'].unique() # посмотрим, какие значения есть в столбце ceiling_height
```

```python
df['ceiling_height'] = df['ceiling_height'].where(~(df['ceiling_height'] >= 20), df['ceiling_height'].apply(lambda x: x / 10)) # разделим на 10 значения, превышающие 20
```

```python
ceiling_height_median = ceiling_height_median.rename(columns={'ceiling_height':'median_ceiling_height'}) # переименуем столбец в median_ceiling_height
display(ceiling_height_median)
```

```python
df = df.merge(ceiling_height_median, on='cityCenters_nearest', how='left') # объединим таблицы так, чтобы в df появился столбец ceiling_height_median
display(df)
```

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df['median_ceiling_height']) # заполним пропущенные значения в столбце ceiling_height медианными значениями
```

```python
df.pop('median_ceiling_height') # удалим столбец ceiling_height_median

```

```python
df['living_rate'] = df['living_area'] / df['total_area'] # вычислим отношение жилплощади к общей площади для каждой квартиры
```

```python
df['locality_name'].unique() # проверим результат замены ё на е 
```

В данных остался ряд дубликатов из-за того, что в некоторых случаях названия сопровождаются категорией населенного пункта. Избавимся от таких дубликатов.  

```python
word_list = ['поселок', 'деревня', 'городского', 'типа', 'село', 'городской', 'садоводческое', 'некоммерческое', 'товарищество', 'коттеджный', 'при', 'железнодорожной', 'станции'] 

clean = r'\b(?:{})\b'.format('|'.join(word_list))

df['clean_locality_name'] = df['locality_name'].str.replace(clean, '') # уберем лишние слова из списка населенных пунктов
```

```python
df['clean_locality_name'].unique()
```

### Этаж квартиры ### 

Проведем категоризацию по этажу 

```python
def floor_category_function(row): # определим функцию, которая будет определять категорию цели получения кредита в зависимости от значения колонки purpose
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый этаж'
    if floor == floors_total:
        return 'последний этаж'
    else: 
        return 'другой'
```

```python
df['floor_category'] = df.apply(floor_category_function, axis=1) # применим функцию floot_category_function
```
